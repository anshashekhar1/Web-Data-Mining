```python
def task3():
    data = dict() # initialize empty dictionary
    
    # add elements to dictionary data
    data['school'] = 'UAlbany'
    data['address'] = '1400 Washington Ave, Albany, NY 12222'
    data['phone'] = '(518) 442-3300'
    
    # loop through dictionary items and print
    for key, value in data.items():
        print(f"{key}: {value}")
    print("\nKeys:")
    print(data.keys())
    print("\nValues:")
    print(data.values())
        
task3()
```

```python
import json

# define dictionary with data
data = {
    'school': 'UAlbany',
    'address': '1400 Washington Ave, Albany, NY 12222',
    'phone': '(518) 442-3300'
}

# convert dictionary to JSON string
json_data = json.dumps(data)

# open file data.json in write mode
with open('data.json', 'w') as file:
    file.write(json_data)

# open file data.json in read mode
with open('data.json', 'r') as file:
    json_string = file.read()
    
# convert JSON string back to dictionary
loaded_data = json.loads(json_string)

print("Items:")
for key, value in loaded_data.items():
    print(f"{key}: {value}")

print("\nKeys:")
print(loaded_data.keys())

print("\nValues:")
print(loaded_data.values())
```

```python
import json

def task5():
    items = [1, 2, 3, 4, 5] # define list of integers
    
    # define dictionary
    data = {
        'school': 'UAlbany',
        'address': '1400 Washington Ave, Albany, NY 12222',
        'phone': '(518) 442-3300'
    }

    objects_to_save = [items, data] # combine the list and dictionary into one list

    # open file task5.data in write mode and save the combine list as JSON
    with open('task5.data', 'w') as file:
        json.dump(objects_to_save, file)

    # open file task5.data in read moad and load JSON data into objects
    with open('task5.data', 'r') as file:
        loaded_objects = json.load(file)

    print("Loaded List:", loaded_objects[0])
    print("Loaded Dictionary:")
    for key, value in loaded_objects[1].items():
        print(f"{key}: {value}")

task5()
```

```python
import json

# open file CrimeReport.txt in read mode
with open('CrimeReport.txt', 'r') as file:
    lines = file.readlines()

tweet_ids = [] # list to store tweet IDs

# loop through each line in the file
for line in lines:
    try:
        tweet = json.loads(line) # try to parse the line as JSON
        tweet_ids.append(tweet['id']) # add the tweet's ID to the list
    except json.JSONDecodeError: # handle JSON decoding errors
        print("Error decoding JSON")
    except KeyError: # handle missing id key in the JSON data
        print("Key 'id' not found in JSON")

# loop through the collected tweet IDs and print
for tweet_id in tweet_ids:
    print(f"Tweet ID: {tweet_id}")
```

```python
import json
import datetime

tweets = [] # initialize list to store tweets

# open CrimeReport.txt to read tweets
with open('CrimeReport.txt', 'r') as file:
    for line in file:
        tweet = json.loads(line) # parse each line from JSON to dictionary
        tweets.append(tweet) # add parsed tweet to the list of tweets

# sort the tweets based on the 'created_at'timestamp
sorted_tweets = sorted(
    tweets, 
    key = lambda item: datetime.datetime.strptime(item['created_at'], '%a %b %d %H:%M:%S +0000 %Y'),
    reverse = True # sort in deceneding order to have most recent tweets first
)

# Write the 10 most recent tweets to task7.data
with open('task7.data', 'w') as f:
    for tweet in sorted_tweets[:10]: # only process the first 10 sorted tweets 
        f.write(json.dumps(tweet) + '\n') # write each tweet as a JSON string
```

```python
import json
import datetime

input_file = 'CrimeReport.txt'
output_dir = 'task8-output/'

# make sure the output dictionary path ends with /
if not output_dir.endswith('/'):
    output_dir += '/'

# reall all lines from input file
with open(input_file, 'r') as file:
    tweets = file.readlines()

# process each tweet in the input file
for line in tweets:
    tweet = json.loads(line) # parse the JSON data from the line
    
    # extract the 'created_at' filed and convert to datatimeo bject
    created_at = datetime.datetime.strptime(tweet['created_at'], '%a %b %d %H:%M:%S +0000 %Y')
    
    # format the filename using the timestamp (year, month, data, hour)
    filename = f"{created_at.strftime('%a-%d-%b-%Y-%H')}.txt"
    filepath = f"{output_dir}{filename}"
    
    # open the file in append mode and write the tweet as JSON
    with open(filepath, 'a') as f:
        f.write(json.dumps(tweet) + '\n')
```
